```python
import ast
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
      L.append(i['name'])
    return L
```

```python
df3['genres']=df3['genres'].apply(convert)
```

```python
df3['keywords']=df3['keywords'].apply(convert)
```

```python
import ast
def convert2(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
    return L
```

```python
df3['cast']=df3['cast'].apply(convert2)
```

```python
import ast
def convert3(obj):
    L =[]
    for i in ast.literal_eval(obj):
      if i['job'] == 'Director':
        L.append(i['name'])
    return L
```

```python
df3['crew']=df3['crew'].apply(convert3)
```

```python
df3['tags']=df3['tags'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
y= PorterStemmer()

def stem(txt):
    li=[]
    for i in txt.split():
        li.append(y.stem(i))
    return " ".join(li)
        
```

```python
df3['tags']=df3['tags'].apply(stem)
```

```python
stem('acting')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors=x.fit_transform(df3['tags']).toarray()
```

```python
vectors[0]
```

```python
x.get_feature_names()
```

```python
stem('happening')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(s[0])), reverse=True , key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    inde= df3[df3['title']==movie].index[0]
    movie_array=s[inde]
    movie_list = sorted(list(enumerate(s[0])), reverse=True , key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df3.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
